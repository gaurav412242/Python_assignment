## CREATING NEW FIELDS .

```python
##if age is > 12  then ageband will be adult  . 
titanic_data.loc[titanic_data['Age']>12 ,'Age_band']= "Adult"
```

now we want to analyse if NAME feature can be engineered to extract Titles and test correlation between Titles and survival , before dropping NAME and PASSENGER ID features .

IN THE FOLLOWING CODE WE WILL EXTRACT TITLE FEATURE USING REGULAR EXPRESSION <BR> THE REGEX PATTERN (\W+\.) MATCHES THE FIRST WORD THAT ENDS WITH . (DOT) CHARACHTER WITHIN THE NAME FEATURE <BR> THE EXPAND = False RETURNS THE DATAFRAME 

```python
titanic_data['Title'] = titanic_data.Name.str.extract('([A-Za-z]+)\.',expand = False) 

pd.crosstab(titanic_data['Title'] ,titanic_data['Sex'] )
```

in the above code we used name.str.extract . then [] means list and inside a group () , it takes one or more occurances with + and takes all the letters from A - z and followed by \. dot <br> expand = false will give dataframe . <br>crosstab will match title with sex . 

REPLACING UNCOMMON AND RARE TITLE WITH RARE . <BR>REPLACING MILE AND MS WITH MISS AND MME WITH MRS 

```python
titanic_data['Title'] = titanic_data['Title'].replace(['Lady','Capt','Col','Countess','Don','Dr','Jonkheer','Major','Rev',
                                                       'Sir'],'Rare')
titanic_data['Title'] = titanic_data['Title'].replace(['Mlle','Ms'],'Miss')

titanic_data['Title'] = titanic_data['Title'].replace('Mme','Mrs')
```

```python
titanic_data[['Title','Survived']].groupby(['Title'],as_index = False).mean()
```


**Insights**

- Most titles band Age groups accurately. For example: Master title has Age mean of 5 years.
- Survival among Title Age bands varies slightly.
- Certain titles mostly survived (Mme, Lady, Sir) or did not (Don, Rev, Jonkheer).

we can convert the categorical title to ordinal 

```python
title_mapping = {'Mr':1 ,'Miss':2 , 'Mrs': 3 , 'Master':4 ,'Rare':5 }
```

```python
titanic_data['Title'] = titanic_data['Title'].map(title_mapping)
```

```python
titanic_data['Title'] = titanic_data['Title'].fillna(0)
```

```python
titanic_data.head()
```



### Decision
#### We decide to retain the new Title feature for model training

NOTE: WHY DO WE CONVERT THE CATEGORICAL TO THE ORDINAL BECAUSE WHEN WE FEED THE DATA TO THE MACHINE LEARNING  MODEL , IT WONT UNDERSTAND THE CATEGORICAL DATA . IT ONLY UNDERSTANDS THE ORDINAL THAT IS NUMERICAL DATA 


### Now we can convert features which contain strings to numerical values. This is required by most model algorithms. Doing so will also help us in achieving the feature completing goal.

- Converting Sex feature to a new feature called Gender where female=1 and male=0.

```python
## we will use  the map function to map male and female and put 1 and 0 
```

```python
titanic_data.head()
```

```python
titanic_data['Sex'] = titanic_data['Sex'].map({'female':1,'male':0}).astype(int)
```

```python
titanic_data.head()
```

AS WE ALREADY HAVE THE TITLE  AND IS CONVERTED TO NUMERICAL FOR MACHINE LEARNING MODEL WE WILL REMOVE THE NAME COLUMN . 

```python
titanic_data.drop('Name' , axis = 1  , inplace = True)
```

```python
titanic_data.head()
```

CONVERTING THE Embarked COLUMN WHICH IS CATEGORICAL INTO NUMERICAL VALUE FOR ML MODEL 

```python
titanic_data['Embarked'] = titanic_data['Embarked'].map({'S':0 , 'C':1 , 'Q':2}).astype(int)
```

```python
titanic_data.head()
```

4.10 **How did total family size affected Survival Count**?

```python
titanic_data['FamilySize'] = titanic_data['SibSp'] + titanic_data['Parch'] + 1

titanic_data[['FamilySize', 'Survived']].groupby(['FamilySize'], as_index=False).mean().sort_values(by='Survived', ascending=False)
```

```python
sns.barplot(x='FamilySize', y='Survived', hue='Sex', data=titanic_data)
```

**Insights**
- Both men and women had a massive drop of survival with a FamilySize over 4. 
- The chance to survive as a man increased with FamilySize until a size of 4
- Men are not likely to Survive with FamilySize 5 and 6
- Big Size Family less likihood of Survival
